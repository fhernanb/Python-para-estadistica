```python
import pandas as pd
from matplotlib import pyplot as plt
```

## Ejemplo de un plot sencillo

```python
x = [1, 2, 3]
y = [1, 4, 9]
plt.plot(x, y)
plt.title("Test plot")
plt.ylabel("Y")
plt.xlabel("X")
plt.show()
```

## Ejemplo de dos plot sencillos

```python
x = [1, 2, 3]
y = [1, 4, 9]
z = [10, 5, 0]
plt.plot(x, y)
plt.plot(x, z)
plt.title("Test plot")
plt.ylabel("Y and Z")
plt.xlabel("X")
plt.legend(["this is y", "this is z"])
plt.show()
```

## Cargando una base de datos sobre países

Para cargar bases de datos se usa la función read_csv

```python
data = pd.read_csv('countries.csv')
```

Queremos comparar el crecimiento de la población de USA y China

```python
us = data[data.country == 'United States']
china = data[data.country == 'China']
```

```python
plt.plot(us.year, us.population / 10**6)
plt.plot(china.year, china.population / 10**6)
plt.legend(["USA", "China"])
plt.ylabel("Population")
plt.xlabel("Year")
plt.show()
```

```python
plt.plot(us.year, us.population / us.population.iloc[0] * 100)
plt.plot(china.year, china.population / china.population.iloc[0] * 100)
plt.legend(["USA", "China"])
plt.ylabel("Population growth (first year = 100)")
plt.xlabel("Year")
plt.show()
```
