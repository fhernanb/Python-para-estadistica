```python
import sympy as sp
```

# Ejemplo 1

```python
x = sp.Symbol("x")
f = input("Ingrese la función en términos de x: \n")

print(f"La función que ingresó es: {f} \n")

f_indefinite = sp.integrate(f, x)
print(f"La integral es: {f_indefinite} \n")

a = float(input("Ingrese el lim inf: "))
b = float(input("Ingrese el lim sup: "))

print("\n")
print(f"El límite inferior es {a}")
print(f"El límite superior es {b} \n")

f_definite = sp.integrate(f, (x, a, b))

print(f"El área bajo la curva de la función {f} es {f_definite}")
```
